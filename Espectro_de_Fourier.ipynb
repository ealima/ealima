# Plotando um sinal com ru√≠do

```python
from scipy.fft import fft, fftfreq, ifft
import matplotlib.pyplot as plt
import numpy as np

N = 1000
Ts = 1.0 / 1000.0
x = Ts*np.arange(N)
ruido_freq =  0.2*np.sin(2*np.pi*x*100)
y = np.sin(10.0 * 2.0*np.pi*x) + 0.5*np.sin(30.0 * 2.0*np.pi*x)
y_ruido = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x) + ruido_freq


plt.rcParams['figure.figsize'] = (25, 5)

plt.subplot(1,3,1)
plt.plot(x, ruido_freq)
plt.subplot(1,3,2)
plt.plot(x, y)
plt.subplot(1,3,3)
plt.plot(x, y_ruido)
```

# Levantando o espectro de magnitude de Fourier

```python
#Transformada Discreta de Fourier

yf = fft(y_ruido)
xf = fftfreq(N, Ts)[:N//2]  # ir√° calcular as frequencias presentes na amostra

import matplotlib.pyplot as plt

plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```

```python
#https://docs.scipy.org/doc/scipy/reference/tutorial/fft.html

N = 20
Ts = 1.0 / 100.0 # Sampling rate, or number of measurements per second

x = Ts*np.arange(N)
y = 1.0*np.sin(230.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x) + 0.2*np.sin(2*np.pi*x*90)
yf = fft(y)
#xf = 1/(N*Ts)*np.arange(N)
xf = fftfreq(N, Ts)[:N//2]  # ir√° calcular as frequencias presentes na amostra

import matplotlib.pyplot as plt

plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```
