```python
from scipy.fft import dct, idct #fast DCT
import matplotlib.pyplot as plt

N = 100 # numero de pontos
t = np.linspace(0,20,N, endpoint=False)
x = np.exp(-t/3)*np.cos(2*t) #amortecimento
y = dct(x, norm='ortho')
window = np.zeros(N)
window[:20] = 1
yr = idct(y*window, norm='ortho')
e=sum(abs(x-yr)**2) / sum(abs(x)**2)

plt.figure(figsize=(25,15))

plt.plot(t, x, '-yo')
plt.plot(t, yr, 'r*')
#plt.plot(t, yr, '-r*')

window = np.zeros(N)
window[:25] = 1 #serÃ¡ que aqui nao era 20 ?
yr = idct(y*window, norm='ortho')
#sum(abs(x-yr)**2) / sum(abs(x)**2)

plt.plot(t, yr, 'g+')
#plt.plot(t, yr, '-g+') liga pontos vermelhos

plt.legend(['x', '$x_{20}$', '$x_{15}$'])
plt.grid()
plt.show()

# window = np.zeros(N) #cria vetor de tamanho 100 com todos preenchidos com 0
# window[:20] = 1 #substituiÃ§Ã£o de todos elementos atÃ© 1
```
