**Transformação na variável independente:**

```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [12, 5]
```

```python
plt.rcParams['figure.figsize'] = [12, 5]

t = np.linspace(-10, 10, 101)
y_original = np.cos(t)*np.exp(-0.1*t)
y_inversa = np.cos(-t)*np.exp(-0.1*-t)

plt.subplot(1,2,1)
plt.axvline(0, 0, color='r')
plt.axhline(0, 0, color='r')
plt.grid()
plt.plot(t, y_original)
plt.subplot(1,2,2)
plt.grid()
plt.axhline(0, 0, color='r')
plt.axvline(0, 0, color='r')
plt.plot(t, y_inversa)
```

**Compressão**

```python
plt.rcParams['figure.figsize'] = [12, 5]

a = 2

t = np.linspace(-10, 10, 101)
y_original = np.sin(t)*np.exp(-0.1*np.abs(t))
y_compressao = np.sin(a*t)*np.exp(-0.1*np.abs(a*t))

plt.subplot(1,2,1)
plt.axvline(0, 0, color='r')
plt.axhline(0, 0, color='r')
plt.grid()
plt.plot(t, y_original)
plt.subplot(1,2,2)
plt.axvline(0, 0, color='r')
plt.axhline(0, 0, color='r')
plt.grid()
plt.plot(t, y_compressao)
```

**Expansão**

```python
plt.rcParams['figure.figsize'] = [12, 5]

a = 0.5 #ver notas de aula 0 < a < 1

t = np.linspace(-10, 10, 101)
y_original = np.sin(t)*np.exp(-0.1*np.abs(t))
y_expansão = np.sin(a*t)*np.exp(-0.1*np.abs(a*t))

plt.subplot(1,2,1)
plt.axvline(0, 0, color='r')
plt.axhline(0, 0, color='r')
plt.grid()
plt.plot(t, y_original)
plt.subplot(1,2,2)
plt.axvline(0, 0, color='r')
plt.axhline(0, 0, color='r')
plt.grid()
plt.plot(t, y_expansão)
```

**Avanço**

```python
plt.rcParams['figure.figsize'] = [12, 5]

a=1
b = -4

t = np.linspace(-10, 10, 101)
y_original = np.sin(t)*np.exp(-0.1*np.abs(t))
y_avanco = np.sin(a*t-b)*np.exp(-0.1*np.abs(a*t-b))

plt.subplot(2,2,1)
plt.axvline(0, 0, color='r')
plt.axhline(0, 0, color='r')
plt.grid()
plt.plot(t, y_original)
plt.subplot(2,2,2)
plt.axvline(0, 0, color='r')
plt.axhline(0, 0, color='r')
plt.grid()
plt.plot(t, y_avanco)
plt.subplot(2,2,3)
plt.axvline(0, 0, color='r')
plt.axhline(0, 0, color='r')
plt.grid()
plt.plot(t, y_original)
plt.plot(t, y_avanco)

```
